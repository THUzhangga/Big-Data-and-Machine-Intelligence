```python
import cv2
#倒入一个OPENCV模块
import matplotlib.pyplot as plt
#绘图模块
%matplotlib inline
#多有图片在行内显示，不会在新cell中显示
```

```python
faceCascade = cv2.CascadeClassifier("/home/techx/opencv/data/haarcascades/haarcascade_frontalface_alt_tree.xml")
#casade classifier是一种分类器，既可以使用Haar，也可以使用LBP特征。
```

```python
Kimi = cv2.imread('Kimi.jpg')
```

```python
print Kimi.shape
print Kimi
```

```python
gray = cv2.cvtColor(Kimi, cv2.COLOR_BGR2GRAY)
img_display = cv2.cvtColor(Kimi, cv2.COLOR_BGR2RGB)
```

```python
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=4,
    minSize=(10, 10),
)
```

```python
len(faces)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img_display, (x, y), (x+w, y+h), (0, 255, 0), 4)
```

```python
plt.imshow(img_display)
```
