# classification using SVM

```python
from sklearn import svm
#SVM(Support Vector Machine)
classifier = svm.SVC(gamma=0.001, C=100.)
```

```python
classifier.fit(digits.data[:-100], digits.target[:-100])
```

```python
y_pred = classifier.predict(digits.data[-100:])
```

```python
y_truth = digits.target[-100:]
```

```python
print(zip(y_pred, y_truth))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# calculate confusion matrix
cm = confusion_matrix(y_pred, y_truth)
```

```python
print(cm)
# wow, the results are pretty good
```
