```python
from sklearn import datasets
```

```python
digits = datasets.load_digits()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
X = pca.fit_transform(digits.data)
```

```python
digits.data.shape
```

```python
print(X.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
s = np.ones((1797,)) * 20
```

```python
plt.scatter(X[:,0], X[:,1], s, digits.target, edgecolors='none')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=10)
```

```python
pca = PCA(n_components=20)
X2 = pca.fit_transform(digits.data)
X2.shape
```

```python
label = km.fit_predict(X2)
```

```python
print(label)
```

```python
plt.scatter(X[:,0], X[:,1], s, label, edgecolors='none')
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(label, digits.target)
```

```python
print(cm)
```
