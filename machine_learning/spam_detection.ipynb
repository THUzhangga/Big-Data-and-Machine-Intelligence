```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(X_text)
#在训练数据中进行fit操作，即建立词表，transform则是进行计数
```

```python
X
print type(X)
```

```python
classifier = SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3, n_iter=5, random_state=42)
```

```python
classifier.fit(X[:-600], y[:-600])
```

```python
y_pred = classifier.predict(X[-600:])
y_truth = y[-600:]
```

```python
from sklearn import metrics
```

```python
print metrics.accuracy_score(y_truth, y_pred)
print metrics.precision_score(y_truth, y_pred)
print metrics.recall_score(y_truth, y_pred)
```

```python
detector = SpamDetector()
detector.set_vectorizer(vectorizer)
detector.set_classifier(classifier)
```
